```python
# Our v1 in previous cell was just a list. Now we change and create v1 as a numpy object.
# Numpy will print v1 as a row and v2 as a column
# row vector
v1 = np.array([ [2,5,4,7] ]) 
print("Shape of v1 :")
print(v1.shape)
print(v1)
# column vector by using Transform function of a row vector
v2 = v1.T
print("Shape of v2 :")
print(v2.shape)
print(v2)
```

```python
# list
v1 = [2,5,4,7] 

# array, no orientation
v2 = np.array(v1)
print(v2.shape)

v3 = v2.T
print(v3.shape)
```

```python

# vecto
v = np.array([1.5,1.5])

# plot it
plt.plot([0,v[0]],[0,v[1]])

# and then plot scaled versions on top
for i in range(10):
    # random scalar
    s = np.random.randn()
    # scaled vector
    sv = s*v
    plt.plot([0,sv[0]],[0,sv[1]])
    
    
plt.grid('on')
plt.axis('square')
plt.axis([-4,4,-4,4])
plt.show()
```
